## Use Tensorflow autograd and gradient descent for linear regression
* create a synthetic dataset with 100 data points

```python
import tensorflow as tf
tf.random.set_seed(0)

X = tf.linspace(-5, 5, 100)
X = tf.cast(X, dtype=tf.float32)
y = 2 * X + tf.random.normal([len(X)], 0., 3.)
```

## Render a scatter plot of the sample data
* remember to import `matplotlib.pyplot as plt` and set `%matplotlib inline`

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X, y);
```

## Initialize the model parameter
* **hint:** don't forget to set `trainable = True`

```python
w = tf.Variable(tf.random.normal([1], 0., 1.), trainable=True)
w
```

## Implement a `forward` method that returns the predicted values

```python
def forward(w, X):
    return w * X
```

## Implement a `loss` method that returns the mean squared error
* **hint:** use the `tf.reduce_mean` method

```python
def loss(y_pred, y):
    return tf.reduce_mean((y_pred - y) ** 2)
```

```python
LEARNING_RATE = 0.03

for _ in range(10):
    with tf.GradientTape() as tape:
      y_pred = forward(w, X)
      mse = loss(y_pred, y)
    
    w_dw = tape.gradient(mse, [w])[0]
    w.assign_sub((LEARNING_RATE * w_dw))
    
    print("MSE: ", mse.numpy().item(), " w: ", w.numpy().item(), " w.grad: ", w_dw.numpy().item())
```

## Render a plot of the sample data with a line based on your model parameter
* use `numpy()` on the tensor generated by your model to render it
* **hint:** reuse your `forward` function to generate the values for the line plot


```python
plt.scatter(X, y)

y_pred = forward(w, X).numpy()
plt.plot(X, y_pred, color='black', linewidth = 4);
```
